```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Findings
1. Base on these data, most cases are in mainland China and Hubei is the epicenter.
2. There are more confirmed cases for male than female.
3. The distribution of case for ages are mostly for adults, the confirmed rate is much higher for elders.

# Import data

lineList_df = pd.read_csv('novel-corona-virus-2019-dataset/COVID19_line_list_data.csv')
lineList_df.info()

```python
confirmed_df = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_confirmed.csv')
deaths_df = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_deaths.csv')
recovered_df = pd.read_csv('novel-corona-virus-2019-dataset/time_series_covid_19_recovered.csv')
confirmed_df
```

```python
confirmed_df = confirmed_df.rename(columns={"Province/State":"state","Country/Region": "country"})
deaths_df = deaths_df.rename(columns={"Province/State":"state","Country/Region": "country"})
recovered_df = recovered_df.rename(columns={"Province/State":"state","Country/Region": "country"})
confirmed_df
```

# Confirmed case in different country


```python
confirmedByCountry_df = confirmed_df.groupby(confirmed_df['country']).sum().sort_values(by = ['2/20/20'], ascending = False)
confirmedByCountry_df.reset_index()
```

# Bar plot of confirmed case in different country

```python
temp = confirmed_df.rename(columns={"2/20/20":"total"})
plt.figure(figsize=(20,6))
plt.bar(temp.country, temp.total)
plt.xticks(rotation=90)
plt.show()
```

# Confirmed case in different province of China

```python
confirmedInChina_df = confirmed_df[confirmed_df['country'] == 'Mainland China']
confirmedInChina_df
```

# Bar plot of confirmed case in different province of China

```python
temp = confirmedInChina_df.rename(columns={"2/20/20":"total"}).sort_values(by = 'total', ascending = False)
plt.figure(figsize=(20,6))
plt.bar(temp.state, temp.total)
plt.xticks(rotation=90)
plt.show()
```

# Number of cases in China

```python
plt.figure(figsize=(20,8))
plt.plot(temp.iloc[:10,4:].T)
plt.legend(temp["state"][:10])
```

# Number of cases in different gender

```python
temp = lineList_df.groupby('gender').count()
plt.figure(figsize=(10,6))
plt.bar(temp.index, temp.id)
plt.xticks(rotation=90)
plt.show()
```

# Number of cases in different age

```python
temp = lineList_df.groupby('age').count()
plt.figure(figsize=(20,6))
plt.bar(temp.index, temp.id)
plt.xticks(rotation=90)
plt.show()
```
