```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', 50)
COVID19_line_list_data = pd.read_csv("COVID19_line_list_data.csv")
COVID19_line_list_data
```

```python
COVID19_line_list_data['death_new'] = np.where(COVID19_line_list_data['death'] != '0', 2, 0)
COVID19_line_list_data['recovered_new'] = np.where(COVID19_line_list_data['recovered'] != '0', 1, 0)
COVID19_line_list_data['outcome'] = COVID19_line_list_data['death_new'] + COVID19_line_list_data['recovered_new']
```

```python
COVID19_line_list_data.groupby('death_new').count()
```

```python
COVID19_line_list_data.groupby('recovered_new').count()
```

```python
COVID19_line_list_data.groupby('outcome').count()
```

```python
df = COVID19_line_list_data.drop(COVID19_line_list_data[COVID19_line_list_data['outcome'] == 0].index)
```

```python
df = df.reset_index(drop=True)
```

```python
df[df['case_in_country'].isnull()]
```

```python
df['case_in_country'].fillna(0,inplace=True)
```

```python
df
```

```python
df.groupby('location').count()
```

```python
df.groupby('country').count()
```

```python
df['is_China'] = np.where(df['country'] == 'China',1,0)
df['is_Japan'] = np.where(df['country'] == 'Japan',1,0)
df['is_Malaysia'] = np.where(df['country'] == 'Malaysia',1,0)
df['is_Singapore'] = np.where(df['country'] == 'Singapore',1,0)
df['is_South_Korea'] = np.where(df['country'] == 'South Korea',1,0)
```

```python
df.groupby('is_South_Korea').count()
```

```python
df
```

```python
df['gender'].replace('male',1,inplace=True)
df['gender'].replace('female',2,inplace=True)
df['gender'].fillna(0,inplace=True)
```

```python
df.groupby('gender').count()
```

```python
df[df['symptom_onset'].isnull()]
```

```python
df
```

```python
import datetime
df['days_to_hospital'] = pd.Series([np.NaN]*len(df))
for i in range(len(df)):
    if df['symptom_onset'].notnull()[i]:
        if df['hosp_visit_date'].notnull()[i]:
            try:
                symptom_date = datetime.datetime.strptime(df['symptom_onset'][i],'%m/%d/%Y')
            except:
                symptom_date = datetime.datetime.strptime(df['symptom_onset'][i],'%m/%d/%y')
            
            try:
                hosp_date = datetime.datetime.strptime(df['hosp_visit_date'][i],'%m/%d/%Y')
            except:
                hosp_date = datetime.datetime.strptime(df['hosp_visit_date'][i],'%m/%d/%y')
            
            df['days_to_hospital'][i] = (hosp_date - symptom_date).days
```

```python
df
```

```python
df[df['exposure_start'].notnull()]
```

```python
df['exposure_time'] = pd.Series([np.NaN]*len(df))
for i in range(len(df)):
    if df['exposure_start'].notnull()[i]:
        if df['exposure_end'].notnull()[i]:
            try:
                exposure_start = datetime.datetime.strptime(df['exposure_start'][i],'%m/%d/%Y')
            except:
                exposure_start = datetime.datetime.strptime(df['exposure_start'][i],'%m/%d/%y')
            
            try:
                exposure_end = datetime.datetime.strptime(df['exposure_end'][i],'%m/%d/%Y')
            except:
                exposure_end = datetime.datetime.strptime(df['exposure_end'][i],'%m/%d/%y')
            
            df['exposure_time'][i] = (exposure_end - exposure_start).days
```

```python
df[df['exposure_time'].notnull()]
```

```python
df_model = pd.DataFrame()
df_model = df[['outcome','case_in_country','is_China','is_Japan','is_Malaysia','is_Singapore','is_South_Korea',
               'gender','age','days_to_hospital','exposure_time','visiting Wuhan']]
```

```python
df_model
```

```python
from sklearn.model_selection import train_test_split
X, y = df_model.iloc[:,1:] , df_model.iloc[:,0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from xgboost import XGBClassifier
m = XGBClassifier(
    max_depth=2,
    gamma=2,
    eta=0.8,
    reg_alpha=0.5,
    reg_lambda=0.5
)
m.fit(X_train, y_train)
m.predict_proba(X_test)[:,1]
```

```python
y_pred = m.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
predict_accuracy_on_test_set = (cm[0,0] + cm[1,1])/(cm[0,0] + cm[1,1]+cm[1,0] + cm[0,1])
predict_accuracy_on_test_set
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred)
```
